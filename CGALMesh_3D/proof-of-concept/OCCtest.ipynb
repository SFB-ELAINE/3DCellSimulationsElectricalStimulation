```python
import netgen.occ as occ
import ngsolve
```

```python
origin = occ.Pnt(0, 0, 0)
sphere1 = occ.Sphere(origin, r=1)
```

```python
sphere2 = sphere1.Move((0.5, 0, 0))
```

```python
from ngsolve.webgui import Draw
```

```python
Draw(sphere1 + sphere2)
```

```python
sphere1cut = sphere1 - sphere2
```

```python
sphere1.mat("C1")
sphere2.mat("C2")
sphere1.bc("M1")
sphere2.bc("M2")
```

```python
glue = occ.Glue([sphere1cut, sphere2])
```

```python
for face in glue.faces:
    print(face.name)
```

```python
occgeo = occ.OCCGeometry(glue)
```

```python
mesh = occgeo.GenerateMesh()
```

```python
Draw(mesh)
```

```python
mm = ngsolve.Mesh(mesh)
```

```python
mm.GetMaterials()
```

```python
mm.GetBoundaries()
```

```python
mesh.FaceDescriptors()
```

```python
for fd in mesh.FaceDescriptors():
    if fd.domin == 1 and fd.domout == 2:
        fd.bcname = "IF"
```

```python
mesh.FaceDescriptors()
```

```python
mesh.GetRegionNames(dim=3)
```

```python
mm = ngsolve.Mesh(mesh)
matcf = mm.MaterialCF({"C1": 1, "C2": 2})
```

```python
Draw(matcf, mm)
```

```python
mesh.Save("test_occ.vol.gz")
```
